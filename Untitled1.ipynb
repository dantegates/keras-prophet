# Define Model

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (
    BatchNormalization,
    Concatenate,
    Dense,
    Embedding,
    Input,
    Layer,
)


def STL(n_models, t_range, seasonal_components=10, output_activation='linear',
        n_outer_layers=None, outer_dimension=None):
    id_input = Input((1,), name='id')
    t_input = Input((1,), name='t')
    
    trends = [
        Trend(n_models, t_range)([id_input, t_input])
    ]
    
    seasonalities = [
        Seasonality(n_models=n_models, n=10, period=period)([id_input, t_input])
        for period in [7., 365.25]
    ]
    W = Concatenate()(trends + seasonalities)

    if n_outer_layers is not None:
        for _ in range(n_outer_layers):
            W = BatchNormalization()(Dense(outer_dimension, activation='relu')(W))
    
    return tf.keras.models.Model(
        inputs=[t_input, id_input],
        outputs=Dense(1, activation=output_activation)(W)
    )


class Seasonality(Layer):
    def __init__(self, n_models, n, period):
        super().__init__()
        self.n_models = n_models
        self.n = n
        self.period = period
        self.a_n = Embedding(input_dim=self.n_models, output_dim=n, input_length=1)
        self.b_n = Embedding(input_dim=self.n_models, output_dim=n, input_length=1)
        
    def __call__(self, inputs):
        ids, t = inputs
        a_n, b_n = self._make_fourier_components(ids)
        period = self.period
        n = (tf.range(self.n, dtype='float') + 1)[None, :]        
        x = (2 * np.pi * tf.multiply(t, n)) / period
        return (
            tf.keras.layers.dot([tf.cos(x), a_n], axes=-1)
            + tf.keras.layers.dot([tf.sin(x), b_n], axes=-1)
        )
    
    def _make_fourier_components(self, ids):
        return self.a_n(ids), self.b_n(ids)


_activations = {
    'identity': lambda x: x,
    # 'sigmoid': cap / (1 + tf.exp(-x))
}


class Trend(Layer):
    def __init__(self, n_models, t_range, n_changepoints=20, changepoint_range=.8, activation='identity'):
        super().__init__()
        self.n_models = n_models
        self.t_range = t_range
        self.n_changepoints = n_changepoints
        self.changepoint_range = changepoint_range
        self.activation = activation
        self.δ = Embedding(input_dim=self.n_models, output_dim=self.n_changepoints, input_length=1)
        self.k = Embedding(input_dim=self.n_models, output_dim=1, input_length=1)
        self.m = Embedding(input_dim=self.n_models, output_dim=1, input_length=1)
        
        
    def __call__(self, inputs):
        ids, t = inputs

        δ = self.δ(ids)
        m = self.m(ids)
        k = self.k(ids)
        changepoints = tf.linspace(
            self.t_range[0],
            int(self.changepoint_range * self.t_range[1]),
            self.n_changepoints + 1)
        s = tf.cast(changepoints, 'float')[1:]

        a_t = tf.convert_to_tensor(t > s)
        A = tf.keras.backend.switch(a_t, tf.ones(self.n_changepoints)[None, :], tf.zeros(self.n_changepoints)[None, :])
        # cast to float and reshape for compatibility with δ
        A = tf.cast(A, 'float')

        trend = (
            tf.keras.layers.multiply([
                tf.keras.layers.add([
                    tf.keras.layers.dot([A, δ], axes=-1),
                    k
                ]),
                t
            ])
        )

        γ = tf.multiply(-s[None, :], δ)
        offset = (
            tf.keras.layers.add([
                tf.keras.layers.dot([A, γ], axes=-1),
                m
            ])
        )
        trend_adjusted = tf.keras.layers.Reshape((1,))(trend + offset)
        
        if self.activation == 'identity':
            outputs = trend_adjusted
        else:
            raise ValueError(f'Unsupported output_activation="{activation}"')

        return outputs
```

# Simulate Data: Multiple Models

```python
import matplotlib.pyplot as plt
import numpy as np

n_items = 100
n_dates = 900
periods = [
    (7, 3),
    (365.25, 20),
]


def generate_seasonalities(t, n, period, order):
    t = np.arange(t)[None, :]
    N = np.arange(order)[:, None]
    a_n = np.random.normal(size=(n, order))
    b_n = np.random.normal(size=(n, order))
    return (
        a_n.dot(np.sin((2 * np.pi * (N * t)) / period))
        + b_n.dot(np.cos((2 * np.pi * (N * t)) / period))
    )


trends = (
    # intercept
    np.random.normal(size=(n_items, 1))
    + np.cumsum(
        # changepoints
        np.random.normal(size=(n_items, n_dates))
        # mask
        * np.random.choice([0, 1], size=(n_items, n_dates), p=[.98, .02]),
        axis=1)
)
trends *= (np.arange(n_dates)[None, :] / 100)
seasonalities = [
    m * generate_seasonalities(n_dates, n_items, P, order)
    for m, (P, order) in zip((5, 25), periods)
]

y = trends + np.sum(seasonalities, axis=0)

fig, axs = plt.subplots(2 + len(seasonalities), figsize=(10, 10))

axs[0].plot(y.T, color='C0', alpha=.4)
_ = axs[1].plot(trends.T, color='C0', alpha=.4)
for ax, s, (P, _) in zip(axs[2:], seasonalities, periods):
    ax.plot(s[:, :round(P)].T, color='C0', alpha=.4)
    
for ax, title in zip(axs, ['Target', 'Trend', *[f'{day} Day Seasonality' for day in periods]]):
    ax.set_title(title)
    
plt.subplots_adjust(top=1.4)
```

```python
import pandas as pd

X = pd.DataFrame({
    'id': np.repeat(np.arange(n_items), n_dates),
    't': np.tile(np.arange(n_dates), n_items),
    'y': y.ravel()
})

Xt = X[['id', 't']].to_dict('series')

model = STL(
    n_models=X.id.nunique(),
    t_range=(X.t.min(), X.t.max()),
    n_outer_layers=5,
    outer_dimension=5,
)
model.compile(loss='mse', optimizer='adam')
model.fit(
    Xt, X.y,
    batch_size=1024,
    epochs=50
)
```

```python
X['p'] = model.predict(Xt)
```

```python
import scipy.stats as ss
import scipy.special as spc


t = np.arange(365 * 5)


def periodic_rbf(x, p, a, m, c):
    return c * np.exp(-.5 * a * ((x % p) - m) ** 2)


product_release = 1# + (t > 500) * .7
table_dropped = 1# - ((t > 1050) & (t < 1100)) * .9

growth = 10 * np.log((t + 90) / 100) * product_release * table_dropped
noise = np.random.normal(size=len(t))

modifier_ix = (t // 365.25).astype(int)
modifier = np.random.normal(0, .7, size=modifier_ix.max() + 1)
yearly_effect1 = periodic_rbf(t, 365.25, .01, 30, 10) - periodic_rbf(t, 365.25, .001, 33, 15)
yearly_effect1 *= abs(modifier[modifier_ix])

yearly_effect2 = periodic_rbf(t, 365.25, .001, 210, 20) / np.log10(t + 10)

weekly_effect = periodic_rbf(t, 7, 1, 5, 5)


y = growth + yearly_effect1 + yearly_effect2 + weekly_effect + noise
y = .03 * (y - y.min()) / (y.max() - y.min()) + .015

plt.plot(y)
```

```python
import scipy.stats as ss


time_to_convert = ss.weibull_min(c=1.2, scale=40)

y_observed = y * time_to_convert.cdf(t.max() - t)

plt.plot(y)
plt.plot(y_observed, ls='--')
```
